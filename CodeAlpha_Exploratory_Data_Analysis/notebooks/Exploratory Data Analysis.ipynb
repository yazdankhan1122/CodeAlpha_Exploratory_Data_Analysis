```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
file_path = r"F:\tableau fff\directory.csv"
df = pd.read_csv(file_path)
print(f"Data loaded successfully! Shape: {df.shape}")
print("First 5 rows:")
df.head()
```

```python
class EDAAnalyzer:
    def __init__(self, df):
        self.df = df.copy()
        self.original_shape = self.df.shape
        self.clean_data()
        print(f"Dataset loaded: {self.original_shape[0]} rows, {self.original_shape[1]} columns")
    
    def clean_data(self):
        self.df_clean = self.df.copy()
        numeric_columns = self.df_clean.select_dtypes(include=[np.number]).columns
        if len(numeric_columns) > 0:
            print(f"Numeric columns found: {list(numeric_columns)}")
        text_columns = self.df_clean.select_dtypes(include=['object']).columns
        if len(text_columns) > 0:
            print(f"Text columns found: {list(text_columns)}")
        return self.df_clean
    
    def ask_questions(self):
        print("MEANINGFUL QUESTIONS TO EXPLORE:")
        questions = [
            "1. What are the basic statistics of numeric columns?",
            "2. Are there any missing values in the dataset?",
            "3. What is the distribution of key numeric variables?",
            "4. Are there correlations between numeric variables?",
            "5. What are the value counts for categorical variables?",
            "6. Are there any outliers in the data?",
            "7. What is the data type distribution?",
            "8. How is the data structured across different categories?"
        ]
        for question in questions:
            print(f"   • {question}")
        return questions
    
    def explore_structure(self):
        print("DATA STRUCTURE ANALYSIS:")
        print("="*50)
        print(f"Dataset Shape: {self.df.shape}")
        print(f"Columns: {list(self.df.columns)}")
        print("Data Types:")
        print(self.df.dtypes)
        print("First 10 rows:")
        display(self.df.head(10))
        print("Descriptive Statistics:")
        display(self.df.describe(include='all'))
        print("Missing Values Analysis:")
        missing_data = self.df.isnull().sum()
        if missing_data.sum() > 0:
            for col, count in missing_data.items():
                if count > 0:
                    print(f"   {col}: {count} missing values ({count/len(self.df)*100:.2f}%)")
        else:
            print("   No missing values found!")
        return self.df.info()
    
    def identify_trends_patterns(self):
        print("TRENDS & PATTERNS ANALYSIS:")
        print("="*50)
        numeric_df = self.df_clean.select_dtypes(include=[np.number])
        if not numeric_df.empty:
            print("Numeric Columns Statistics:")
            for col in numeric_df.columns:
                col_stats = self.df_clean[col].describe()
                print(f"   {col}:")
                for stat, value in col_stats.items():
                    print(f"      {stat}: {value:.2f}")
                Q1 = self.df_clean[col].quantile(0.25)
                Q3 = self.df_clean[col].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR
                outliers = self.df_clean[(self.df_clean[col] < lower_bound) | (self.df_clean[col] > upper_bound)]
                print(f"      Outliers detected: {len(outliers)}")
        categorical_df = self.df_clean.select_dtypes(include=['object'])
        if not categorical_df.empty:
            print("Categorical Columns Patterns:")
            for col in categorical_df.columns:
                if self.df_clean[col].nunique() < 20:
                    value_counts = self.df_clean[col].value_counts()
                    print(f"   {col}:")
                    for value, count in value_counts.head().items():
                        print(f"      {value}: {count}")
        return None
    
    def test_hypotheses(self):
        print("HYPOTHESIS TESTING:")
        print("="*50)
        numeric_df = self.df_clean.select_dtypes(include=[np.number])
        if len(numeric_df.columns) >= 2:
            print("Correlation Analysis:")
            correlation_matrix = numeric_df.corr()
            display(correlation_matrix)
            high_corr_pairs = []
            for i in range(len(correlation_matrix.columns)):
                for j in range(i+1, len(correlation_matrix.columns)):
                    if abs(correlation_matrix.iloc[i, j]) > 0.5:
                        high_corr_pairs.append(
                            (correlation_matrix.columns[i], correlation_matrix.columns[j], 
                             correlation_matrix.iloc[i, j])
                        )
            if high_corr_pairs:
                print("High Correlation Pairs (|r| > 0.5):")
                for col1, col2, corr in high_corr_pairs:
                    print(f"   {col1} vs {col2}: {corr:.3f}")
            else:
                print("   No strong correlations found")
        if not numeric_df.empty:
            print("Normality Tests:")
            for col in numeric_df.columns:
                if self.df_clean[col].nunique() > 10:
                    stat, p_value = stats.normaltest(self.df_clean[col].dropna())
                    print(f"   {col}: p-value = {p_value:.4f}")
                    if p_value > 0.05:
                        print(f"      → Normally distributed")
                    else:
                        print(f"      → NOT normally distributed")
    
    def detect_data_issues(self):
        print("DATA QUALITY ISSUES:")
        print("="*50)
        issues = []
        missing_data = self.df.isnull().sum()
        if missing_data.sum() > 0:
            issues.append(f"Missing values found: {missing_data[missing_data > 0].to_dict()}")
        duplicates = self.df.duplicated().sum()
        if duplicates > 0:
            issues.append(f"Duplicate rows found: {duplicates}")
        numeric_cols = self.df_clean.select_dtypes(include=[np.number]).columns
        for col in numeric_cols:
            if self.df_clean[col].nunique() > 2:
                skewness = self.df_clean[col].skew()
                if abs(skewness) > 2:
                    issues.append(f"High skewness in {col}: {skewness:.2f}")
        categorical_cols = self.df_clean.select_dtypes(include=['object']).columns
        for col in categorical_cols:
            unique_count = self.df_clean[col].nunique()
            if unique_count == len(self.df_clean):
                issues.append(f"Column '{col}' might be an ID (all unique values)")
        if issues:
            print("Potential issues found:")
            for issue in issues:
                print(f"   • {issue}")
        else:
            print("No major data quality issues detected")
        return issues
    
    def create_visualizations(self):
        print("CREATING VISUALIZATIONS...")
        numeric_cols = self.df_clean.select_dtypes(include=[np.number]).columns
        categorical_cols = self.df_clean.select_dtypes(include=['object']).columns
        
        if len(numeric_cols) > 0:
            n_numeric = len(numeric_cols)
            n_cols = min(3, n_numeric)
            n_rows = (n_numeric + n_cols - 1) // n_cols
            fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
            if n_numeric == 1:
                axes = np.array([axes])
            if n_rows == 1:
                axes = axes.reshape(1, -1)
            axes = axes.flatten()
            for i, col in enumerate(numeric_cols):
                if i < len(axes):
                    axes[i].hist(self.df_clean[col].dropna(), bins=20, alpha=0.7, color='skyblue', edgecolor='black')
                    axes[i].set_title(f'Distribution of {col}')
                    axes[i].set_xlabel(col)
                    axes[i].set_ylabel('Frequency')
                    axes[i].grid(True, alpha=0.3)
            for j in range(i+1, len(axes)):
                axes[j].set_visible(False)
            plt.tight_layout()
            plt.show()
        
        if len(categorical_cols) > 0:
            categorical_for_plot = [col for col in categorical_cols if self.df_clean[col].nunique() < 10]
            if categorical_for_plot:
                n_cat = len(categorical_for_plot)
                n_cols = min(2, n_cat)
                n_rows = (n_cat + n_cols - 1) // n_cols
                fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
                if n_cat == 1:
                    axes = np.array([axes])
                if n_rows == 1:
                    axes = axes.reshape(1, -1)
                axes = axes.flatten()
                for i, col in enumerate(categorical_for_plot):
                    if i < len(axes):
                        value_counts = self.df_clean[col].value_counts()
                        colors = plt.cm.Set3(np.linspace(0, 1, len(value_counts)))
                        bars = axes[i].bar(value_counts.index, value_counts.values, color=colors)
                        axes[i].set_title(f'Distribution of {col}')
                        axes[i].set_xlabel(col)
                        axes[i].set_ylabel('Count')
                        axes[i].tick_params(axis='x', rotation=45)
                for j in range(i+1, len(axes)):
                    axes[j].set_visible(False)
                plt.tight_layout()
                plt.show()
        
        if len(numeric_cols) >= 2:
            print("Correlation Heatmap:")
            correlation_matrix = self.df_clean[numeric_cols].corr()
            plt.figure(figsize=(10, 8))
            sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
            plt.title('Correlation Heatmap')
            plt.tight_layout()
            plt.show()
        
        if len(numeric_cols) > 0:
            print("Boxplots for Outlier Detection:")
            n_box = len(numeric_cols)
            n_cols = min(3, n_box)
            n_rows = (n_box + n_cols - 1) // n_cols
            fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
            if n_box == 1:
                axes = np.array([axes])
            if n_rows == 1:
                axes = axes.reshape(1, -1)
            axes = axes.flatten()
            for i, col in enumerate(numeric_cols):
                if i < len(axes):
                    sns.boxplot(y=self.df_clean[col], ax=axes[i])
                    axes[i].set_title(f'Boxplot of {col}')
                    axes[i].set_ylabel(col)
            for j in range(i+1, len(axes)):
                axes[j].set_visible(False)
            plt.tight_layout()
            plt.show()
    
    def generate_comprehensive_report(self):
        print("STARTING COMPREHENSIVE EDA ANALYSIS...")
        print("="*60)
        self.ask_questions()
        self.explore_structure()
        self.identify_trends_patterns()
        self.test_hypotheses()
        issues = self.detect_data_issues()
        self.create_visualizations()
        print("COMPREHENSIVE EDA REPORT SUMMARY")
        print("="*30)
        print(f"Dataset Overview:")
        print(f"   • Size: {self.original_shape[0]} rows × {self.original_shape[1]} columns")
        numeric_cols = self.df_clean.select_dtypes(include=[np.number]).columns
        categorical_cols = self.df_clean.select_dtypes(include=['object']).columns
        print(f"   • Numeric columns: {len(numeric_cols)}")
        print(f"   • Categorical columns: {len(categorical_cols)}")
        print(f"   • Data issues found: {len(issues)}")
        missing_total = self.df.isnull().sum().sum()
        print(f"   • Total missing values: {missing_total}")
        print("Key Insights:")
        if len(numeric_cols) > 0:
            print("   • Numeric variables analyzed for distributions and outliers")
        if len(categorical_cols) > 0:
            print("   • Categorical variables examined for value distributions")
        if len(numeric_cols) >= 2:
            print("   • Correlation analysis performed between numeric variables")
        print("Next Steps:")
        next_steps = [
            "Address any data quality issues identified",
            "Proceed with targeted analysis based on findings",
            "Consider data transformation if needed",
            "Prepare for advanced modeling if applicable"
        ]
        for step in next_steps:
            print(f"   • {step}")

eda = EDAAnalyzer(df)
eda.generate_comprehensive_report()

print("VERIFICATION: Cleaned Data Sample")
print("="*40)
display(eda.df_clean.head(10))

print("EDA COMPLETED SUCCESSFULLY!")
```
